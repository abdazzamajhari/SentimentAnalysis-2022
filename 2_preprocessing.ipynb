```python
import pandas as pd
```

```python
df = pd.read_csv("tweet.csv", names=['date', 'tweet', 'user'])
df
```

```python
import re

def cleanTxt(text):
    text = re.sub('@[A-Za-z0â€“9]+', '', text) # Menghapus @mentions
    text = re.sub('#', '', text) # Menghapus '#' hash tag
    text = re.sub('_', '', text) # Menghapus '_' underscore
    text = re.sub(':', '', text) # Menghapus ':' 
    text = re.sub('RT[\s]+', '', text) # Menghapus RT
    text = re.sub('https?\/\/\S+', '', text) # Menghapus hyperlink
    # https://www.adamsmith.haus/python/answers/how-to-remove-single-quotes-from-a-string-in-python
    text = re.sub("b'", '', text) # Menghapus b'
    text = re.sub("'", '', text) # Menghapus ' --> quotes
    return text
```

```python
df['tweet'] = df['tweet'].apply(cleanTxt)
df
```

```python
# https://stackoverflow.com/questions/35657918/series-object-has-no-attribute-strip

df['tweet'] = df['tweet'].str.replace(r'\\n', '')
df
```

# Mencari Tweet Bot
Parameter Tweet Bot memiliki tulisan atau kata-kata yang mirip (tidak identik), buzzer dari sisi pemerintah/oposisi, tweet yang duplikat, dan bahasan yang kurang nyambung<br>
https://towardsdatascience.com/finding-and-removing-duplicate-rows-in-pandas-dataframe-c6117668631f

```python
import matplotlib.pyplot as plt
import numpy as np

# Count the number of duplicates
tweet_bot = len(df.loc[df['tweet'].duplicated() == True])
# Count the number of non-duplicates
tweet_normal = len(df.loc[~df['tweet'].duplicated()])
labels = 'Bot', 'Normal'
sizes = np.array([tweet_bot, tweet_normal])
colors = ['lightskyblue', 'pink']
explode= (0, 0.5)
def absolute_value(val):
    a  = np.round(val/100.*sizes.sum(), 0)

    a=  str(round(val,2))+"%"+'\n'+str(a) +" data"
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, explode=explode,shadow=True)

plt.axis('equal')
plt.title("Data Proportion")
plt.legend()
plt.show()
# plt.savefig("data split.jpg",dpi=300)
```

# Analisis Data
Tweet dimulai pada tanggal 11 Maret 2022 - 19 Maret 2022 terkait kelangkaan minyak goreng<br>
Terdapat 3 class, yaitu normal(mendukung pemerintah), negatif(kritik pemerintah), dan bot

```python
df = pd.read_csv(r"tweet2-fix.csv", encoding='latin-1', sep=';')
df
```

```python
# Remove column name 'A'
df = df.drop(['id', 'date', 'user'], axis = 1)
```

```python
# Considering certain columns for dropping duplicates
df = df.drop_duplicates(subset=['tweet'])
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import numpy as np

tweet_normal = len(df.loc[df['class'] == 'normal'])
tweet_negatif = len(df.loc[df['class'] == 'negatif'])
tweet_bot = len(df.loc[df['class'] == 'bot'])

labels = 'Normal', 'Negatif', 'Bot'
sizes = np.array([tweet_normal, tweet_negatif, tweet_bot])
colors = ['lightskyblue', 'pink', 'red']
explode= (0.25, 0.25, 0.25)
def absolute_value(val):
    a  = np.round(val/100.*sizes.sum(), 0)

    a=  str(round(val,2))+"%"+'\n'+str(a) +" data"
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, explode=explode,shadow=True)

plt.axis('equal')
plt.title("Data Proportion")
plt.legend()
plt.show()
# plt.savefig("data split.jpg",dpi=300)
```
